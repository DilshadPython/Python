```python
import matplotlib.pyplot as plt
import numpy as np

'''
Plottingx and y points
The plot() function is used to draw the points(markers) in a diagram
By default, the plot() function draws a line from point to point

The Parameter 1 is an array containing the points on the x-axis
THe Parameter 2 is an array containing the points on the y-axis

We draw a line x-axis is horizantol distance(0, 10) and y-axis is vertical distance(0, 150)

'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

xline = np.array([0, 10])
yline = np.array([0, 150])

# draw the lines
plt.plot(xline, yline, color='red')
plt.show()
```

```python
xline = np.array([0, 10])
yline = np.array([0, 150])

yline = np.array([10, 0])
xline = np.array([150, 0])

# draw the lines
plt.plot(xline, yline, color='red')
plt.plot(yline, xline, color='green')
plt.show()
```

```python
xline = np.array([2, 20])
yline = np.array([1, 18])

# lower o not zero as string
plt.plot(xline, yline, 'o', color='red')
plt.show()
```

# Markers
The keyword argtument marker to emphasize each point with a specified marker:
Mark each point with a cricle

```python
yline = np.array([5, 15, 10, 25, 4])

plt.plot(yline, marker = 'o', color='blue')
plt.show()
```

```python
plt.plot(yline, marker='*')
plt.show()
```

```python
plt.plot(yline, marker='X', color='red')
plt.show()
```
