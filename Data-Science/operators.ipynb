## Sorted manually

```python
nums = [3, 7, 0, -5, -1, 9, 2, 3, -3, 10, -7, 0]

for i in range(len(nums)):
    num = nums[i]
    index = i
    cop = i
    for x in range(i, len(nums)):
        if nums[x] < num:
            num = nums[x]
            index = cop
        cop += 1
    newlst = nums[i]
    nums[i] = num
    nums[index] = newlst
print(nums)
```

```python
def get_avarage_marks(desk):
    avarge_mark = {}
    for i in desk:
        lst = desk[i]
        num = 0
        for nums in lst:
            num += int(nums)
        avarge_mark[i] = num/len(lst)
    return avarge_mark
```

```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.impute import Simplelmputer 

data_frame = pd.read_csv('nonSentinelTestsDetections.csv')

data_frame.drop(['SNo', 'Last Update'], axis=1, inplace=True)
data_frame.rename(columns={'ObervationDate':'Date', 'Province/State':'State', 'Country/Region':'Country'}, inplace=True)

data_frame['Date'] = pd.to_datetime(data_frame['Date'])

imputer = SimpleImputer(strategy='constant')

data_frame_1 = pd.DataFrame(imputer.fit_transform(data_frame), columns=data_frame.columns)
data_frame_2 = data_frame_1.groupby(['Date', 'Country'])[['Date', 'Country', 'Confirmed', 'Deaths', 'Recovered']].sum.reset_index()

data_frame_3 = data_frame_2.groupby(['Date', 'Country'])[['Date', 'Country', 'Confirmed', 'Deaths', 'Recovered']].sum.reset_index()

data_frame_3.head(20)
```
