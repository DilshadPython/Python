```python
import pandas as pd
```

```python
students = {
    'std_id': [12345, 12346, 12347, 12348],
    'fname': ['Dilshad', 'Adam', 'Chris', 'Azad'],
    'lname': ['Abdulla', 'George', 'Smith', 'Abdulla'],
    'email': ['dilshad.a73@email.com', 'adam_george@email.com', 'chris_smith@email.com', 'azad_abdulla@email.com'],
    'city': ['Berlin', 'Roma', 'Paris', 'London']
}
```

```python
df = pd.DataFrame(students)
```

```python
df
```

```python
df.sort_values(by='fname')
```

```python
df.sort_values(by='lname')
```

```python
# ascending use for reverse if is True means following by lname but if is False not reverese it by lname
df.sort_values(by='fname', ascending=False)
```

```python
df.sort_values(by='fname', ascending=True)
```

```python
# this is sorted by lname first when all lname done than follow fname of course if ascending=False
df.sort_values(by=['lname', 'fname'], ascending=False)
```

```python
# it not sorted data by lname or fname because ascending=True
df.sort_values(by=['lname', 'fname'], ascending=True)
```

```python
df.sort_values(by=['lname', 'fname'], ascending=False)
```

```python
# this is sort all lname first
df.sort_values(by=['lname', 'fname'])
```

```python
# this is sort all lname first
df.sort_values(by=['city', 'lname', 'fname'])
```

```python
# the sort doesn't work here because ascending=True
df.sort_values(by=['city', 'lname', 'fname'], ascending=True)
```

```python
# the sort follow city first and it work here because ascending=False
df.sort_values(by=['city', 'lname', 'fname'], ascending=True)
```

```python
df.sort_values(by=['city', 'lname'], ascending=[False, True])
```

```python
df.sort_values(by=['city', 'lname'], ascending=[True, False])
```

```python
df.sort_values(by=['city', 'lname'], ascending=[True, True])
```

```python
df.sort_values(by=['city', 'lname'], ascending=[False, False])
```

```python
# inplace=True doesn't display the result at the same line we have to run df in the next line nothing change in the sort
df.sort_values(by=['city', 'lname'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
# inplace=False display the result at the sametime this line is run nothing change in the sort
df.sort_values(by=['city', 'lname'], ascending=[False, True], inplace=False)
```

```python
df.sort_values(by=['city', 'lname'], ascending=[True, True], inplace=False)
```

```python
df.sort_values(by=['city', 'lname'], ascending=[True, True], inplace=True)
```
