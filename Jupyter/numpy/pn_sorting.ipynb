```python
import pandas as pd
```

```python
students = {
    'std_id': [12345, 12346, 12347, 12348],
    'fname': ['Dilshad', 'Adam', 'Chris', 'Azad'],
    'lname': ['Abdulla', 'George', 'Smith', 'Abdulla'],
    'email': ['dilshad.a73@email.com', 'adam_george@email.com', 'chris_smith@email.com', 'azad_abdulla@email.com'],
    'city': ['Berlin', 'Roma', 'Paris', 'London']
}
```

```python
df = pd.DataFrame(students)
```

```python
df
```

```python
# sorting by last name
df.sort_values(by='lname')
```

```python
# sorting by first name
df.sort_values(by='fname')
```

```python
df.sort_values(by='std_id')
```

```python
# Desending order like reverse
df.sort_values(by='lname', ascending=False)
```

```python
# Desending order like reverse
df.sort_values(by='fname', ascending=False)
```

```python
# Desending order like reverse
df.sort_values(by='std_id', ascending=False)
```

```python
# Desending order like reverse
df.sort_values(by='email', ascending=False)
```

```python
# Change ascending order to True it will display by order 
df.sort_values(by='email', ascending=True)
```

```python
# Desending order like reverse
df.sort_values(by=['lname', 'fname'], ascending=False)
```

```python
# Desending order like reverse
df.sort_values(by=['lname', 'fname'], ascending=True)
```

```python
# Desending order like reverse
df.sort_values(by=['lname', 'fname', 'city'], ascending=True)
```

```python
# Desending order like reverse
df.sort_values(by=['city', 'lname', 'fname'], ascending=True)
```

```python
# Desending order like reverse
df.sort_values(by=['lname', 'fname'], ascending=False)
```
